```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

features = pd.read_csv("Features data set.csv")
sales = pd.read_csv("sales data-set.csv")
stores = pd.read_csv("stores data-set.csv")

print("Features shape:", features.shape)
print("Sales shape:", sales.shape)
print("Stores shape:", stores.shape)

print("\n--- FEATURES HEAD ---")
display(features.head())
print("\n--- SALES HEAD ---")
display(sales.head())
print("\n--- STORES HEAD ---")
display(stores.head())

# COMPLETENESS
print("\n--- Missing Values: FEATURES ---")
display(features.isnull().sum())

print("\n--- Missing Values: SALES ---")
display(sales.isnull().sum())

print("\n--- Missing Values: STORES ---")
display(stores.isnull().sum())

# Check if all sales.Store exist in stores.Store
missing_store_keys = set(sales['Store']) - set(stores['Store'])
print("Stores referenced in SALES but missing in STORES:", missing_store_keys)

# Check (Store, Date) integrity between sales and features
sales_pairs = set(zip(sales['Store'], sales['Date']))
features_pairs = set(zip(features['Store'], features['Date']))
missing_feature_pairs = sales_pairs - features_pairs

print("Sales (Store,Date) pairs missing in FEATURES:", len(missing_feature_pairs))

# VALIDITY
# Negative or zero weekly sales
invalid_sales = sales[sales["Weekly_Sales"] < 0]
print("Negative Weekly Sales rows:", len(invalid_sales))

# Check for invalid temperature values
if "Temperature" in features.columns:
    invalid_temp = features[features["Temperature"] < -30]
    print("Invalid Temperature rows:", len(invalid_temp))

# Check for invalid CPI values
if "CPI" in features.columns:
    invalid_cpi = features[features["CPI"] <= 0]
    print("Invalid CPI rows:", len(invalid_cpi))

# CONSISTENCY
sales['Date'] = pd.to_datetime(sales['Date'], errors='coerce')
features['Date'] = pd.to_datetime(features['Date'], errors='coerce')

print("Invalid date conversion in SALES:", sales['Date'].isnull().sum())
print("Invalid date conversion in FEATURES:", features['Date'].isnull().sum())

# DUPLICATES
sales_dups = sales.duplicated().sum()
features_dups = features.duplicated().sum()
stores_dups = stores.duplicated().sum()

print("Duplicate rows in SALES:", sales_dups)
print("Duplicate rows in FEATURES:", features_dups)
print("Duplicate rows in STORES:", stores_dups)

# SUMMARY
print("\n--- SALES Summary ---")
display(sales.describe())

print("\n--- FEATURES Summary ---")
display(features.describe())

print("\n--- STORES Summary ---")
display(stores.describe())

# Visualization
plt.figure(figsize=(8,4))
sales['Weekly_Sales'].hist(bins=50)
plt.title("Distribution of Weekly Sales")
plt.xlabel("Weekly Sales")
plt.ylabel("Frequency")
plt.show()
```
